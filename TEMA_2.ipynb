Rezolvarea sistemelor folosind descompunerea LU a unei matrici.

Se considera sistemul :

\begin{cases} x + 2y + z + t = 0  \\ x + 4y - z + 7t = 20 \\ 
4x + 9y + 5z  + 11t = 18 \\ x + 6z + 4t = 1 \end{cases}

Astfel, sistemul se va memora folosind matricea sistemului $A$ si $b$, coloana termenilor liberi:

\begin{equation}
A =
\begin{bmatrix}
1 & 2 & 1 & 1 \\
1 & 4 & -1 & 7 \\
4 & 9 & 5 & 11 \\
1 & 0 & 6 & 4 
\end{bmatrix}
\end{equation}

iar, 

\begin{equation}
b =
\begin{bmatrix}
0  \\
20  \\
18  \\
1
\end{bmatrix}
\end{equation}

Pasii spre rezolvarea sistemului:

$\textbf{Pasul 1.}$ Gasim descompunerea $LU$ a matricei $A$, a sistemului. 

$L$ este o matrice subdiagonala (lower), deci are elemente doar sub diagonala principala. Deasupra diagonalei principale vom avea doar zerouri.

$U$ este o matrice supradiagonala (upper), deci are elemente doar deasupra diagonalei principale. Sub diagonala principala va avea doar zerouri.

Matricele $L, U$ trebuie sa verifice relatia $A = L * U$.

\begin{equation}
\begin{bmatrix}
1 & 2 & 1 & 1 \\
1 & 4 & -1 & 7 \\
4 & 9 & 5 & 11 \\
1 & 0 & 6 & 4 
\end{bmatrix}
=
 \begin{bmatrix}
 a & 0 & 0 & 0 \\
 b & c & 0 & 0 \\
 d & e & f & 0 \\
 g & h & i & j
\end{bmatrix}
*
\begin{bmatrix}
    1 & x & y & z \\
    0 & 1 & t & u \\
    0 & 0 & 1 & v \\
    0 & 0 & 0 & 1
\end{bmatrix} 
\end{equation}

Idee de gasire a celor doua matrici $L$ si $U$ este una simpla: Vom incerca efectiv sa gasim necunoscutele $a, b, c, d, e, f, g, h, i, j$ pentru $L$ si $x, y, z, t, u, v$ pentru $U$.

Deoarece relatia pe care o cunoastem este ca $A = L*U$ (adica o inmultire de matrici, iar inmultirea a doua matrici, $L$ si $U$ in cazul nostru, presupune inmultirea unei linii din $L$ cu o coloana din $U$, deducem ca trebuie sa aflam pe rand cate o linie din $L$ si cate o coloana din $U$. Vom incepe cu cea care are cele mai putine necunoscute, deci vom afla mai intai o coloana din $U$ si apoi o linie din $L$, treptat). 

Pe prima coloana din $U$, 
\begin{equation}
U_{1} = 
\begin{bmatrix}
    1  \\
    0  \\
    0  \\
    0 
\end{bmatrix}
\end{equation}
nu avem nicio necunoscuta, deci o consideram aflata.

Aflam acum prima linie din $L$: 

\begin{equation}
L_{1} = 
\begin{bmatrix}
    a & 0 & 0 & 0\\ 
\end{bmatrix}
\end{equation}

Pe $a$ il vom afla folosind faptul ca $L_{1} * U_{1} = a_{11}$, unde $L_{1}$ este linia 1 din $L$, $U_{1}$ este coloana 1 din $U$, iar $a_{11}$ este elementul de la linia 1, coloana 1 din $A$.

Relatia $L_{1} * U_{1} = a_{11}$ inseamna ca $a*1 + 0*0 + 0*0 + 0*0 = 1$, deci $a=1$. 

Deci prima data am aflat $U_{1} = [1,  0, 0, 0] ^ {T}$, iar $L_{1} = 
[1, 0, 0, 0]$, unde $X^{T}$ reprezinta transpusa matricei $X$. 

La urmatoarea iteratie, vom afla $U_2$, si $L_2$ din $U$, respectiv $L$(coloana 2 si apoi linia 2 din $U$ si apoi din $L$).
\begin{equation}
U_{2} = 
\begin{bmatrix}
    x \\ 1 \\ 0 \\ 0 
\end{bmatrix}
\end{equation}

Necunoscuta este $x$. Vom folosi elementul corespunzator lui $x$ din $A$: $L_{1} * U_{2} = a_{12}$, deci $ 1*x + 0*1 + 0+0 + 0*0 = 2$, deci $x = 2$.

Aflam acum $L_2 = [b, c, 0, 0]$. Mai intai pe $b$, apoi pe $c$. 

Pentru $b$, vom folosi elementul corespunzator din $A$. 

$a_{21} = L_{2} * U_{1}$, deci $1 = b*1 + c*0 + 0*0 + 0*0$. Rezulta $b=1$.

Pentru $c$, vom avea: $a_{22} = L_{2} * U_{2}$, asadar, $4 = b*x + c*1 + 0*0 + 0*0$, si cum $b, x$ sunt deja aflate, rezulta ca $b = 2$.

La urmatoarea iteratie, aflam $U_3$ si apoi $L_3$. Pentru $U_3$, aflam, ca mai sus, elementele de la 
\begin{matrix}
     \text{linia 1, coloana 3 (din U), folosind $L_{1} * U_{3} = a_{13}$}\\
     \text{linia 2, coloana 3 (din U), folosind $L_{2} * U_{3} = a_{23}$}\\
     \text{la linia 3, coloana 3 avem deja 1}
\end{matrix}

 Pentru $L_3$, aflam, ca mai sus, elementele de la 
\begin{matrix}
     \text{linia 3, coloana 1 (din L), folosind $L_{3} * U_{1} = a_{31}$}\\
     \text{linia 3, coloana 2 (din L), folosind $L_{3} * U_{2} = a_{32}$}\\
     \text{linia 3, coloana 3 (din L), folosind $L_{3} * U_{3} = a_{33}$}
\end{matrix}

Procedam la fel si pentru $L_{4} \text{ si } U_{4}$ si aflam $L$ si $U$.

Algoritmul ar arata cam asa:


```python
A = [[1, 2, 1, 1],
    [1, 4, -1, 7],
    [4, 9, 5, 11],
    [1, 0, 6, 4]]

L = [[0] * 4] * 4
U = [[0] * 4] * 4

def getElementOnColumn(j, i, A, L, U):
    # ------ to do ------
    # a_0j = L[0][0] * U[0][j] + L[0][1] * U[1][j] + ...  = L[0][0] * U[0][j]
    # a_1j = L[1][0] * U[0][j] + L[1][1] * U[1][j] + L[1][2] * U[2][j] + ... = L[1][0] * U[0][j] + L[1][1] * U[1][j]
    # ....
    # a_tj = L[t][0] * U[0][j] + L[t][1] * U[1][j] + ...  + L[t][t - 1] * U[t - 1][j] + L[t][t] * U[t][j]
    #       -----------------am aflat deja elementele din U pana aici --------il mai aflu pe U[t][j]-----
    # U[t][j] = (a_tj - (L[t][0] * U[0][j] + L[t][1] * U[1][j] + ...  + L[t][t - 1] * U[t - 1][j])) / L[t][t]

    return 0

def getElementOnLine(i, j, A, L, U):
    # ------ to do ------
    # a_i0 = L[i][0] * U[0][0] + L[i][1] * U[1][0] + ... = L[i][0] * U[0][0]
    # a_i1 = L[i][0] * U[0][1] + L[i][1] * U[1][1] + L[i][1] * U[1][2] + ... = L[i][0] * U[0][1] + L[i][1] * U[1][1]
    # ...
    # a_it = L[i][0] * U[0][t] + L[i][1] * U[1][t] + ... + L[i][t - 1] * U[t - 1][t] + L[i][t] * U[t][t]
    # L[i][t] = a_it - (L[i][0] * U[0][t] + L[i][1] * U[1][t] + ... + L[i][t - 1] * U[t - 1][t])
    return 0

def getColumnU(i, A, L, U):
    for j in range(i):
        U[j][i] = getElementOnColumn(j, i, A, L, U)

def getLineL(i, A, L, U):
    for j in range(i + 1):
        L[i][j] = getElementOnLine(i, j, A, L, U)
        
dimA = 4

for i in range(dimA):
    getColumnU(i, A, L, U)
    getLineL(i, A, L, U)
    
```
